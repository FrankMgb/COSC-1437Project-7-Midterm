#Chi-squared test

1. Individual getting easily upset are more prone to have depression
Individual who are anxious are more protne to have depression
Similar statement for the demographic (maritial status, gender, age)


Hythothesis testing 1
Null hypthosis (H0): There is no association between getting easily upset and depression
Alternative hypothesis (H1): Individual getting easily upset are more prone to have depression

Chi-square: it finds the association between two factors

```python
# prompt: perform chi square test between columns "depressed" and "getting_upset"

from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(data['depressed'], data['getting_upset'])

# Perform the chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies table:")
print(expected)

# Interpret the results
alpha = 0.05  # Significance level
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between getting easily upset and depression.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between getting easily upset and depression.")
```

Cramer'v: this finds how strong is the association?

```python
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

results = []

for col in yes_columns:
    table = pd.crosstab(data[col], data['depressed'])
    v = cramers_v(table)
    results.append((col, v))

# Sort by Cramér’s V (effect size)
sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

print("Top contributing features by Cramér’s V:")
for col, v in sorted_results[:100]:
    print(f"{col}: Cramér’s V = {v:.4f}")
```
